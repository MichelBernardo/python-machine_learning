```python
import torch
from diffusers import StableDiffusionXLPipeline
import matplotlib.pyplot as plt
```

```python
pretrained_model = "stabilityai/stable-diffusion-xl-base-1.0"

prompt = '''
Ultra-realistic cinematic football match scene,
professional football player in mid-action kicking the ball with full force, dynamic body movement,
intense facial expression, packed stadium in the background, roaring crowd, dramatic stadium floodlights,
rain particles in the air, motion blur on the ball, sharp focus on the player, detailed jersey fabric,
sweat and dirt on skin, realistic anatomy, hyper-detailed muscles, global illumination, volumetric lighting,
cinematic color grading, HDR, shallow depth of field, 70–85mm lens, low-angle shot, epic sports photography,
8k detail, masterpiece, award-winning sports photography, photorealistic
'''

negative_prompt ='''
low quality, worst quality, blurry, out of focus, bad anatomy, deformed body, extra limbs, extra fingers,
missing fingers, bad hands, distorted face, cross-eye, flat lighting, overexposed, underexposed, cartoon,
anime, illustration, painting, plastic skin, waxy skin, unrealistic proportions, duplicated players, cloned faces,
watermark, text, logo
'''

num_inference_steps = 50
guidance_scale = 8
height = 1024
width = 1024
seed = 1235
```

```python
generator = torch.manual_seed(seed)

pipeline = StableDiffusionXLPipeline.from_pretrained(
    pretrained_model,
    dtype=torch.float16,
    use_safetensors=True,
    variant="fp16"
)

pipeline.to('cuda')
```

```python
image = pipeline(
    prompt=prompt,
    negative_prompt=negative_prompt,
    height=height,
    width=width,
    num_inferences_steps=num_inference_steps,
    guidance_scale=guidance_scale,
    generator=generator
).images[0]
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```
