```python
from sklearn import datasets
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score
```

```python
dataset = datasets.load_iris()
dataset
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
kmeans.fit(dataset.data)

print(kmeans.labels_)

result = confusion_matrix(dataset.target, kmeans.labels_)
print(result)

accuracy = accuracy_score(kmeans.labels_, dataset.target)
precision = precision_score(kmeans.labels_, dataset.target, average='weighted')
print(f'Accuracy: {accuracy}\nPrecision: {precision}')
```

```python
dbscan = DBSCAN(eps=0.5, min_samples=3)
dbscan_labels = dbscan.fit_predict(dataset.data)
print(dbscan_labels)

result = confusion_matrix(dataset.target, dbscan.labels_)
print(result)

accuracy = accuracy_score(dbscan_labels, dataset.target)
precision = precision_score(dbscan_labels, dataset.target, average='weighted')
print(f'Accuracy: {accuracy}\nPrecision: {precision}')
```

```python
agglomerative = AgglomerativeClustering(n_clusters=3)
agglomerative_labels = agglomerative.fit_predict(dataset.data)
print(agglomerative_labels)

result = confusion_matrix(dataset.target, agglomerative_labels)
print(result)

accuracy = accuracy_score(agglomerative_labels, dataset.target)
precision = precision_score(agglomerative_labels, dataset.target, average='weighted')
print(f'Accuracy: {accuracy}\nPrecision: {precision}')
```
