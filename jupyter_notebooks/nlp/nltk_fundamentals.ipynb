```python
import nltk

from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer, LancasterStemmer

import string
```

```python
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

nltk.download('tagsets')
nltk.download('tagsets_json')
nltk.download("averaged_perceptron_tagger_eng")

nltk.download("maxent_ne_chunker_tab")
nltk.download("wordnet")
nltk.download("words")
```

```python
Texto = '''
No princípio era a Palavra, e a Palavra estava com Deus e a Palavra era Deus. 
Aquele que é a Palavra estava com Deus no princípio.
Todas as coisas foram feitas por intermédio dele; sem ele, nada do que existe teria sido feito. 
Nele estava a vida, e esta era a luz dos homens. A luz brilha nas trevas, e as trevas não a derrotaram.
Surgiu um homem enviado por Deus, chamado João. Ele veio como testemunha para testificar acerca da luz, 
a fim de que por meio dele todos os homens cressem. Ele próprio não era a luz, mas veio como testemunha da luz. 
Estava chegando ao mundo a verdadeira luz, que ilumina todos os homens.
Aquele que é a Palavra estava no mundo, e o mundo foi feito por intermédio dele, mas o mundo não o reconheceu.
'''
```

```python
print(Texto)
```

```python
sentences = sent_tokenize(Texto)
print(sentences)
print(len(sentences))
```

```python
tokens = word_tokenize(Texto)
print(tokens)
print(len(tokens))
```

```python
stop_words = stopwords.words('portuguese')
print(stop_words)
```

```python
tokens_without_stop_words = [w for w in tokens if w not in stop_words]
print(tokens)
print(tokens_without_stop_words)
```

```python
print(string.punctuation)
```

```python
tokens_without_stop_words_and_punctuation = [w for w in tokens_without_stop_words if w not in string.punctuation]
print(tokens_without_stop_words)
print(tokens_without_stop_words_and_punctuation)
```

```python
frequency = nltk.FreqDist(tokens_without_stop_words_and_punctuation)
frequency
```

```python
most_commons_tokens = frequency.most_common(5)
print(most_commons_tokens)
```

```python
stemmer1 = PorterStemmer()
stem1 = [stemmer1.stem(w) for w in tokens_without_stop_words_and_punctuation]

print(tokens_without_stop_words_and_punctuation)
print(stem1)
```

```python
nltk.help.upenn_tagset()
```

```python
pos_tags = nltk.pos_tag(tokens_without_stop_words_and_punctuation)
print(pos_tags)
```

```python
lemmatizer = WordNetLemmatizer()
result = [lemmatizer.lemmatize(w) for w in tokens_without_stop_words_and_punctuation]
print(tokens_without_stop_words_and_punctuation)
print(result)
```

```python
test = nltk.ne_chunk(pos_tags)
print(test)
```
